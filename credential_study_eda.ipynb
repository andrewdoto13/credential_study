# Credential Study

## Import Libraries

```python
import numpy as np
import pandas as pd
import sklearn
import re

np.random.seed(13)
print(sklearn.__version__)
```

## Import Data

```python
df = pd.read_csv("JFP Credential Data - ProQuest v2.csv")
```

```python
df.head()
```

## Extract Credentials

```python
df["Credentials"] = df.Authors.apply(lambda x: re.findall(r"(CFP|PhD|CFA|JD|CPA|CLU|ChFC|LLM|AIF|CRC)", str(x))).apply(lambda x: np.nan if len(x) == 0 else x)
```

```python
clean_df = df[["Journal", "Title", "Authors", "Abstract", "pubdate", "year", "Credentials"]]
```

```python
clean_df.loc[clean_df.Abstract.isna()][0:20]
```
